```python
departments = {}
for _, row in objects.iterrows():
    if np.isnan(row['department_id']):
        if np.isnan(row['department_name']):
            continue
        else:
            raise 'ID is false, but Name is not'
    if row['department_id'] in departments:
        assert row['department_name'] == departments[row['department_id']]
    else:
        departments[row['department_id']] = row['department_name']
```

```python
def save_departments(departments):
    pd.DataFrame({'id': departments.keys(), 'name': departments.values()}).to_sql('departments', con(), index=False)
```

```python
sportzones_table = pd.DataFrame({
    'id': sportzones.keys(),
    'name': [item['name'] for item in sportzones.values()],
    'type': [item['type'] for item in sportzones.values()]
})
sportzones_table.head()
```
